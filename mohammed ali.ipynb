```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\samue\\Downloads\\Muhammad Ali Boxing.csv")
df
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
df['Age(years)'] = df['Age(in years)']+(df['Age(additional days)']/365)
```

```python
df = df.drop(['Age(in years)', 'Age(additional days)'], axis=1)
```

```python
tko_wins = df[(df['Result'] == 'Win') & (df['Type'] == 'TKO')]

print(tko_wins)
```

```python
outcome_counts = df['Result'].value_counts()
print(outcome_counts)
```

```python
opponent_count = df['Opponent'].value_counts().head(10)
print(opponent_count)
```

```python
# Fights by year
df['Year'] = df['Date']
fights_per_year = df.groupby('Year').size()
print(fights_per_year)
```

```python
location_outcomes = df.groupby(['Location', 'Result']).size().unstack()
print(location_outcomes)
```

```python
import matplotlib.pyplot as plt

fights_per_year.plot(kind='bar', figsize=(20, 6), title="Fights Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Fights")
plt.tight_layout()
plt.show()
```

```python
outcome_counts.plot(kind='pie', autopct='%1.1f%%', title="Fight Outcomes")
plt.ylabel('')
plt.show()
```
